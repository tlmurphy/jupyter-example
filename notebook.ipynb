```python
# Read csv from URL
import pandas as pd
csv_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
headers = ["SepalLength", "SepalWidth", "PetalLength", "PetalWidth", "Name"]
df = pd.read_csv(csv_url, names = headers)
df.dtypes
```

```python
# Plot the data using andres_curves
from pandas.plotting import andrews_curves
andrews_curves(df, "Name")
```

```python
# Get the mean of the data
df.mean(numeric_only=True)
```

```python
# Get the median of the data
df.median(numeric_only=True)
```

```python
# Show only Virginica with PetalWidth < 2 and SepalLength > 7 
df.query("Name == 'Iris-virginica' and PetalWidth < 2 and SepalLength > 7")
```
